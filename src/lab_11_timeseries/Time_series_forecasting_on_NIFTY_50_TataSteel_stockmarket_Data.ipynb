# Simple Moving Averages:

This function is used to prepare data for time series forecasting. It takes a dataset and generates features and labels for training a model.

Here's a breakdown of the function:

- `dataset`: This is the input data, which is a sequence of values (like a time series of stock prices).
- `start_index` and `end_index`: These are the indices in the dataset that define the range of data to use for generating the features and labels.
- `window_size`: This is the size of the sliding window to use for generating the features. For each window of data in the dataset, a feature and a label will be generated.

The function works by sliding a window of size `window_size` across the dataset from `start_index` to `end_index`. For each window of data, it appends the data in the window to the `features` list and the data at the next time step to the `labels` list.

The `range(i, i+window_size)` creates a range of indices for the current window of data. The `dataset[l]` selects the data at these indices from the dataset.

The `l[-1]+1` gets the index of the next time step after the current window. The `dataset[l[-1]+1]` selects the data at this index from the dataset.

Finally, the function returns the features and labels as numpy arrays.

The resulting features and labels can be used to train a model to predict the value at the next time step given a window of previous values.

```python
def prepare_data(dataset,start_index,end_index,window_size):
    labels=[]
    features=[]
    for i in range(start_index,end_index-window_size):
        l=range(i,i+window_size)
        features.append(dataset[l])
        labels.append(dataset[l[-1]+1])

    return np.array(features),np.array(labels)
```

```python
x_train,y_train=prepare_data(uni_data,0,4000,5)
x_test,y_test=prepare_data(uni_data,4000,5141,5)
```

```python
# Return mean of the input data
def moving_averages_model(input_data):
    return np.mean(input_data)
```

```python
# Calculate the moving average of X_train
y_pred=[]
for i in range(len(x_train)):
    y_pred.append(moving_averages_model(x_train[i]))
```

```python
#Calculate mean squared error between actual and predicted values
from sklearn.metrics import mean_squared_error

error=mean_squared_error(y_train,np.array(y_pred))
```

Below code is using a moving averages model to predict time series data and then calculating the mean squared error of the predictions. It's doing this for different window sizes to see how the window size affects the prediction error.

Here's a breakdown of the code:

1. `window=[3,4,5,6,7,8,9,10]`: This is a list of window sizes that will be used for the moving averages model. The window size is the number of previous time steps that are used to calculate the moving average.

2. `train_loss=[]` and `test_loss=[]`: These are lists that will store the mean squared error of the predictions for the training and testing data, respectively.

3. The outer `for` loop (`for i in window:`) iterates over each window size in the `window` list.

4. `x_train,y_train=prepare_data(uni_data,0,4000,i)` and `x_test,y_test=prepare_data(uni_data,4000,5141,i)`: These lines call the `prepare_data` function to generate the training and testing data. The `prepare_data` function is assumed to take a dataset, a start index, an end index, and a window size, and return a set of features and labels for each window of data in the dataset.

5. The next two inner `for` loops iterate over each window of data in the training and testing datasets, respectively. For each window of data, the moving averages model is used to make a prediction, which is appended to the `y_train_pred` and `y_test_pred` lists.

6. `train_loss.append(mean_squared_error(y_train,np.array(y_train_pred)))` and `test_loss.append(mean_squared_error(y_test,np.array(y_test_pred)))`: These lines calculate the mean squared error of the predictions for the training and testing data, respectively, and append the errors to the `train_loss` and `test_loss` lists.

At the end of the code, the `train_loss` and `test_loss` lists will contain the mean squared error of the predictions for each window size in the `window` list. This can be used to determine the optimal window size for the moving averages model.

```python
window=[3,4,5,6,7,8,9,10]  # window size
train_loss=[]
test_loss=[]
for i in window:
    x_train,y_train=prepare_data(uni_data,0,4000,i)
    x_test,y_test=prepare_data(uni_data,4000,5141,i)
    y_train_pred=[]
    y_test_pred=[]
    for j in range(len(x_train)):
        y_train_pred.append(moving_averages_model(x_train[j]))
    for j in range(len(x_test)):
        y_test_pred.append(moving_averages_model(x_test[j]))
    train_loss.append(mean_squared_error(y_train,np.array(y_train_pred)))
    test_loss.append(mean_squared_error(y_test,np.array(y_test_pred)))

```

```python
plt.rcParams["figure.figsize"] = [10, 6]
```

```python
plt.plot(train_loss)
plt.plot(test_loss)
plt.xlim(3,10)
plt.xlabel('k-values')
plt.ylabel('losses')
```

# Univariate time-series forecasting:LSTM'S

```python
x_train,y_train=prepare_data(uni_data,0,4000,10)
x_test,y_test=prepare_data(uni_data,4000,5141,10)
```

```python
x_train_mean=np.mean(x_train,axis=0)
x_train_std=np.std(x_train,axis=0)
x_train = (x_train-x_train_mean)/x_train_std
x_test = (x_test-x_train_mean)/x_train_std
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## Model Architecture:

This code is defining a simple LSTM (Long Short-Term Memory) model using the Keras API in TensorFlow. Here's a breakdown:

1. `from tensorflow.keras.layers import Dense,LSTM,Input`: This line imports the necessary layers for building the model. `Dense` is a fully connected layer, `LSTM` is a type of recurrent neural network layer, and `Input` is used to instantiate a Keras tensor.

2. `from tensorflow.keras.models import Model`: This line imports the `Model` class from Keras. A model is a way of organizing layers into a network with training and inference features.

3. `input1=Input(shape=(10,1))`: This line creates an input layer for the model. The `shape` argument specifies the shape of the input data, which in this case is `(10, 1)`. This means the model expects input data with 10 time steps and 1 feature per time step.

4. `x=LSTM(8,activation='relu')(input1)`: This line adds an LSTM layer to the model with 8 units (or neurons). The `activation` argument specifies the activation function to use, which in this case is the ReLU (Rectified Linear Unit) function.

5. `x=Dense(16,activation='relu')(x)`: This line adds a Dense layer to the model with 16 units and the ReLU activation function.

6. `x=Dense(1)(x)`: This line adds another Dense layer to the model with 1 unit. Since no activation function is specified, a linear activation function is used by default.

7. `model=Model(inputs=input1,outputs=x)`: This line creates the model. The `inputs` argument specifies the input(s) of the model, and the `outputs` argument specifies the output(s) of the model.

The resulting model takes sequences of 10 values as input and outputs a single value. This could be used, for example, for time series forecasting where you want to predict the next value based on the previous 10 values.

```python
from tensorflow.keras.layers import Dense,LSTM,Input
from tensorflow.keras.models import Model

input1=Input(shape=(10,1))
x=LSTM(8,activation='relu')(input1)
x=Dense(16,activation='relu')(x)
x=Dense(1)(x)
model=Model(inputs=input1,outputs=x)

model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='mae')
```

## Model Training:

```python
model.fit(x_train,y_train,batch_size=8, epochs=10, verbose=1, validation_data=(x_test,y_test))
```

# Predictions And Plot Results:

```python
y_test_pred=model.predict(x_test)
plt.plot(y_test,label='actual')
plt.plot(y_test_pred,label='predicted')
plt.xlabel('time')
plt.ylabel('price')
plt.legend()
plt.title('predictions')
```

# Multi-variate & single-step forecasting:LSTM'S

This code is preparing data for a multivariate time series forecasting problem. It's creating sequences of data (features) and corresponding labels for training and testing a model.

Here's a breakdown of the function `prepare_data_mul`:

- `dataset`: This is the input data, which is a 2D array where each row is a time step and each column is a variable in the time series.
- `start_index` and `end_index`: These are the indices in the dataset that define the range of data to use for generating the features and labels.
- `window_size`: This is the size of the sliding window to use for generating the features. For each window of data in the dataset, a feature and a label will be generated.

The function works by sliding a window of size `window_size` across the dataset from `start_index` to `end_index`. For each window of data, it appends the data in the window to the `features` list and the data at the next time step to the `labels` list.

The `range(i, i+window_size)` creates a range of indices for the current window of data. The `dataset[l]` selects the data at these indices from the dataset.

The `l[-1]+1` gets the index of the next time step after the current window. The `dataset[l[-1]+1][-1]` selects the last variable at this time step from the dataset.

Finally, the function returns the features and labels as numpy arrays.

The lines after the function definition convert a pandas DataFrame `data` to a numpy array and call the `prepare_data_mul` function to generate training and testing data. The training data uses the first 4000 time steps of the dataset and the testing data uses the rest of the dataset. The window size is set to 10.

```python
def prepare_data_mul(dataset,start_index,end_index,window_size):
    labels=[]
    features=[]
    for i in range(start_index,end_index-window_size):
        l=range(i,i+window_size)
        features.append(dataset[l])
        labels.append(dataset[l[-1]+1][-1])

    return np.array(features),np.array(labels)

data1=data.values
x_train,y_train=prepare_data_mul(data1,0,4000,10)
x_test,y_test=prepare_data_mul(data1,4000,5141,10)
```

#### Normailize the data

```python
x_train_mean=np.mean(x_train,axis=0)
x_train_std=np.std(x_train,axis=0)
x_train = (x_train-x_train_mean)/x_train_std
x_test = (x_test-x_train_mean)/x_train_std
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## Model Architecture:

This code is defining a deep learning model using the Keras API in TensorFlow. The model is a type of Recurrent Neural Network (RNN) called Long Short-Term Memory (LSTM), which is commonly used for sequence data like time series or text.

Here's a breakdown:

1. `from tensorflow.keras.layers import Dense,LSTM,Input`: This line imports the necessary layers for building the model. `Dense` is a fully connected layer, `LSTM` is a type of recurrent neural network layer, and `Input` is used to instantiate a Keras tensor.

2. `from tensorflow.keras.models import Model`: This line imports the `Model` class from Keras. A model is a way of organizing layers into a network with training and inference features.

3. `input1=Input(shape=(10,7))`: This line creates an input layer for the model. The `shape` argument specifies the shape of the input data, which in this case is `(10, 7)`. This means the model expects input data with 10 time steps and 7 features per time step.

4. `x=LSTM(32,activation='relu',return_sequences=True)(input1)`: This line adds an LSTM layer to the model with 32 units (or neurons). The `activation` argument specifies the activation function to use, which in this case is the ReLU (Rectified Linear Unit) function. The `return_sequences=True` argument means that this LSTM layer will return the full sequence of outputs (one output vector per time step), which is necessary when stacking LSTM layers.

5. `x=LSTM(16,activation='relu')(x)`: This line adds another LSTM layer with 16 units and the ReLU activation function. By default, this LSTM layer will only return the output for the last time step.

6. `x=Dense(16,activation='relu')(x)`: This line adds a Dense layer to the model with 16 units and the ReLU activation function.

7. `x=Dense(1)(x)`: This line adds another Dense layer to the model with 1 unit. Since no activation function is specified, a linear activation function is used by default.

8. `model=Model(inputs=input1,outputs=x)`: This line creates the model. The `inputs` argument specifies the input(s) of the model, and the `outputs` argument specifies the output(s) of the model.

9. `model.compile(optimizer='rmsprop', loss='mae')`: This line compiles the model with the RMSprop optimizer and the Mean Absolute Error (MAE) loss function. This prepares the model for training.

The resulting model takes sequences of 10 values with 7 features each as input and outputs a single value. This could be used, for example, for multivariate time series forecasting where you want to predict the next value based on the previous 10 values and 7 features.

```python
from tensorflow.keras.layers import Dense,LSTM,Input
from tensorflow.keras.models import Model

input1=Input(shape=(10,7))
x=LSTM(32,activation='relu',return_sequences=True)(input1)
x=LSTM(16,activation='relu')(x)
x=Dense(16,activation='relu')(x)
x=Dense(1)(x)
model=Model(inputs=input1,outputs=x)
model.compile(optimizer='rmsprop', loss='mae')
model.summary()
```

## Model Training:

```python
model.fit(x_train,y_train,batch_size=8, epochs=10, verbose=1, validation_data=(x_test,y_test))
```

# Predictions And Plot Results:

```python
y_test_pred=model.predict(x_test)
plt.plot(y_test,label='actual')
plt.plot(y_test_pred,label='predicted')
plt.xlabel('time')
plt.ylabel('price')
plt.legend()
plt.title('predictions')
```
